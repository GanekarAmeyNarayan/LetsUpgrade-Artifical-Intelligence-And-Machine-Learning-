```python
x = data[['Age', 'Experience', 'Income', 'Family', 'CCAvg', 'Education','Mortgage', 'Securities Account', 'CD Account', 'Online', 'CreditCard']]
y = data['Personal Loan']
```

```python
print(len(x))
print(len(y))
```

```python
x1 = sm.add_constant(x)
```

```python
Logistic = sm.Logit(y,x1)
```

```python
model = Logistic.fit()
```

```python
model.summary()
```

1. p value for Age, Experience, mortage are > 0.5 that implies they are least significant in deciding outcome

2. p value for Income, Family, CCAvg, Education, Securities Account, CD Account, online, CreditCard are close to 0 that implies they are highly significant in decinding outcome

3. Now we can find probalbility equation and after substituting respective values we can decide whethar a peson is eligible to get loan or not

4. If probability is < 0.5 Then they have low chances of getting loan approved.

5. If probability is > 0.5 Then they have high chances of getting loan approved.
